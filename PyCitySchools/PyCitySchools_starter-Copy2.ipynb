```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

```

```python
#print dimensions of school_data
print("school_data dimensions: {}".format(school_data.shape))
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete 
```

```python
#Calculate total number of schools
count_schools = len(school_data_complete["school_name"].unique())
count_schools
```

```python
#Calculate total number of student
count_students = len(school_data_complete["Student ID"].unique())
count_students = "{:,}".format(count_students)
count_students
```

```python
#Calculate total budget using dataframe school_data
total_budget = school_data["budget"].sum()
total_budget = "${:,.2f}".format(total_budget)
total_budget

```

```python
#Calculate the average math score by using mean
avg_math_score = round((student_data["math_score"].mean()), 6)
avg_math_score
```

```python
#Calculate the average reading score by using mean
avg_reading_score = round((student_data["reading_score"].mean()), 5)
avg_reading_score
```

```python
#Calculate the percentage of students with a passing math and reading score (70 or greater)
student_data["passing_math"] = student_data["math_score"]>=70
student_data["passing_reading"] = student_data["reading_score"]>=70
pct_passing_math = round(((student_data["passing_math"]).mean())*100,6)
pct_passing_reading = round(((student_data["passing_reading"]).mean())*100,6)
#pct_passing_math = "{:.6f}".format(pct_passing_math)
#pct_passing_reading = "{:.6f}".format(pct_passing_reading)
pct_passing_math
                         
```

```python
pct_passing_reading
```

```python
#Calculate Overall Passing students
student_data["overall_passing"] = student_data["passing_math"] & student_data["passing_reading"]
pct_overall_passing = round(((student_data["overall_passing"]).mean())*100,6)
pct_overall_passing
```
